```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# PPG  
loaded_npy = np.load('downloads/dataset.npy')

#  (CSV )
labels = pd.read_csv('downloads/dataset_caseid.csv') 

#  54000, 9  
sampling_rate = 100  # 1/100  (100Hz)
duration = 540  #   540 (9)
time = np.arange(0, duration, 1/sampling_rate)  #  (0~540)

plt.figure(figsize=(12, 8))

#  3   (  3 )
for i in range(3):  
    if labels.iloc[i, 0] == 1:  #   
        plt.plot(time, loaded_npy[i, :], label=f'Arrhythmic Patient {i+1}', color='red', alpha=0.7)

plt.title("arrhythmia PPG")
plt.xlabel("Time (seconds)")
plt.ylabel("PPG Value")
plt.legend()
plt.grid(True)
plt.show()

#   
plt.figure(figsize=(12, 8))
for i in range(3):  #   3 
    if labels.iloc[i, 0] == 0:  #   
        plt.plot(time, loaded_npy[i, :], label=f'Normal Patient {i+1}', color='green', alpha=0.7)

plt.title("Normal PPG")
plt.xlabel("Time (seconds)")
plt.ylabel("PPG Value")
plt.legend()
plt.grid(True)
plt.show()

#        ( )
plt.figure(figsize=(12, 8))
for i in range(3):  #  3  
    color = 'red' if labels.iloc[i, 0] == 1 else 'green'
    label = f'Patient {i+1} (Arrhythmic)' if labels.iloc[i, 0] == 1 else f'Patient {i+1} (Normal)'
    plt.plot(time, loaded_npy[i, :], label=label, color=color, alpha=0.7)

plt.title("Multiple Patients (Arrhythmic vs Normal)")
plt.xlabel("Time (seconds)")
plt.ylabel("PPG Value")
plt.legend()
plt.grid(True)
plt.show()
```
