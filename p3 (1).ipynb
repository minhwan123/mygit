```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# PPG 데이터 
loaded_npy = np.load('downloads/dataset.npy')

# 레이블 데이터(CSV 파일)
labels = pd.read_csv('downloads/dataset_caseid.csv') 

# 환자당 54000, 9분 동안의 데이터
sampling_rate = 100  # 1/100초 샘플링 (100Hz)
duration = 540  # 데이터 길이는 540초 (9분)
time = np.arange(0, duration, 1/sampling_rate)  # 시간 축(0~540초)

plt.figure(figsize=(12, 8))

# 첫 3명의 환자만 시각화 (예시로 첫 3명의 데이터)
for i in range(3):  
    if labels.iloc[i, 0] == 1:  # 부정맥 환자일 경우
        plt.plot(time, loaded_npy[i, :], label=f'Arrhythmic Patient {i+1}', color='red', alpha=0.7)

plt.title("arrhythmia PPG")
plt.xlabel("Time (seconds)")
plt.ylabel("PPG Value")
plt.legend()
plt.grid(True)
plt.show()

# 정상 환자만 시각화
plt.figure(figsize=(12, 8))
for i in range(3):  # 예시로 첫 3명의 데이터
    if labels.iloc[i, 0] == 0:  # 정상 환자일 경우
        plt.plot(time, loaded_npy[i, :], label=f'Normal Patient {i+1}', color='green', alpha=0.7)

plt.title("Normal PPG")
plt.xlabel("Time (seconds)")
plt.ylabel("PPG Value")
plt.legend()
plt.grid(True)
plt.show()

# 여러 환자의 데이터를 색상에 따라 구분하여 시각화 (부정맥과 정상)
plt.figure(figsize=(12, 8))
for i in range(3):  # 첫 3명의 환자 데이터
    color = 'red' if labels.iloc[i, 0] == 1 else 'green'
    label = f'Patient {i+1} (Arrhythmic)' if labels.iloc[i, 0] == 1 else f'Patient {i+1} (Normal)'
    plt.plot(time, loaded_npy[i, :], label=label, color=color, alpha=0.7)

plt.title("Multiple Patients (Arrhythmic vs Normal)")
plt.xlabel("Time (seconds)")
plt.ylabel("PPG Value")
plt.legend()
plt.grid(True)
plt.show()
```
